```python
#Q2 scatter is a type of graph in which data is represented by dots.

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.plot(x,y)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
plt.show()
```
